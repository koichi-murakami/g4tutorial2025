```python
import pandas as pd
import matplotlib.pyplot as plt

# CSVファイルを読み込む
data = pd.read_csv("Bragg.csv")
x = data.iloc[:, 0]  # 第一カラム（x）
y = data.iloc[:, 1]  # 第二カラム（y）

# グラフの作成
plt.figure(figsize=(12, 6))


# yのヒストグラム
plt.subplot(1, 2, 1)
plt.hist(y, bins=100, color='green', alpha=0.7)
plt.title("Histogram of eDep")
plt.xlabel("eDep")
plt.ylabel("Frequency")

# xとyの散布図 (scatter plot)
plt.subplot(1, 2, 2)
plt.scatter(x, y, color='red', alpha=0.6, s=1)
plt.title("Scatter Plot of copyNo vs eDep")
plt.xlabel("copyNo")
plt.ylabel("eDep")

# グラフのレイアウトを調整して表示
plt.tight_layout()
plt.show()
```
