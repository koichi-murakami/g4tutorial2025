```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('P11_MT.csv')

# Extract the two columns
#column1 = data[1]
#column2 = data[2]
column1 = data.iloc[:, 0]  # First column
column2 = data.iloc[:, 1]  # Second column
# Create the plot
plt.figure(figsize=(10, 5))

# Histogram for Column1
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.hist(column1, bins=20, color='blue', alpha=0.7, label='eDepo')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Column1')
plt.legend()

# Histogram for Column2
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.hist(column2, bins=20, color='green', alpha=0.7, label='TotStepLen')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Column2')
plt.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
