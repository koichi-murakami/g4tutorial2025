```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
data = pd.read_csv('P11_MT.csv')

# Extract the two columns
#column1 = data[1]
#column2 = data[2]
column1 = data.iloc[:, 0]  # First column
column2 = data.iloc[:, 1]  # Second column
# Create the plot
plt.figure(figsize=(10, 5))

# Histogram for Column1
plt.subplot(1, 3, 1)  # 1 row, 2 columns, 1st subplot
plt.hist(column1, bins=100, color='blue', alpha=0.7, label='eDepo')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Column1')
plt.legend()

# Histogram for Column2
plt.subplot(1, 3, 2)  # 1 row, 2 columns, 2nd subplot
plt.hist(column2, bins=100, color='green', alpha=0.7, label='TotStepLen')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Column2')
plt.legend()

# Scatter plot for the two columns
plt.subplot(1, 3, 3)  # 1 row, 3 columns, 3rd subplot
plt.scatter(column1, column2, color='purple', alpha=0.7, label='eDep vs TotLength', s=1)
plt.xlabel('First Column')
plt.ylabel('Second Column')
plt.title('Scatter Plot of Two Columns')
plt.legend()

# Adjust layout and show the plot
plt.tight_layout(pad=3.0)
plt.show()
```
