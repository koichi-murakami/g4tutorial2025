```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
df = pd.read_csv("P09_Output.csv", header=None)
#data
mu=df[0].mean()
sigma = df[0].std()
xmin = mu-3.*sigma
xmax = mu + 3.*sigma
#gaussian
x = np.linspace(xmin, xmax, 200)
y = norm.pdf(x,loc = mu, scale = sigma)
#superpose data and graph
fig, ax = plt.subplots() 
ax.hist(df[0], density=True, range=[xmin, xmax], bins=200, label="Data")
ax.plot(x, y, 'r', linewidth=1, label="Gaussian")
#minimum cosmetics
ax.grid(True)
ax.legend(loc='upper right')
ax.set_xlabel("Energy(MeV)")
ax.set_ylabel("normalized frequency")
title = "Result: mu = %.2f, std = %.2f" % (mu, sigma)
plt.title(title)
#Et voila
plt.savefig("P09_electron1GeV.png")
plt.show()


#data.head()

#fig = plt.figure(figsize=(12,8))
#ax1 = fig.add_subplot(2,2,1)


#ax1.grid(True)
#ax1.set_xlabel("E deposit(MeV)")
#ax1.set_ylabel("Step length (mm)")
#ax1.set_title("electron into BGO")
##ax1.set_ylim(8000, 12000.)
##ax1.set_xlim(800, 1000)
#ax1.scatter(df[0], df[1], s=1, marker= '.')

#ax2 = fig.add_subplot(2,2,2)
#ax2.set_ylabel("Step length (mm)")
#ax2.set_xscale('log')
#ax2.hist(df[1], bins=200,orientation='horizontal')  #, range=[8000., 12000.])

#ax3 = fig.add_subplot(2,2,3)
#ax3.set_yscale('log')
#ax3.set_xlabel("Energy Deposit (MeV)")
#ax3.hist(df[0], bins=200) #, range=[800,  1000])

#plt.show()
```
